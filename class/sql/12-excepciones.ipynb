<h1>EXCEPCIÓN</h1>

- Solo los registros que están en una tabla pero no en la otra
- Solo se incluyen los registros que están en la tabla izquierda pero no en la derecha

```python
select * from world.cities where world.cities.name = 'Abidjan'
```

```python
select * from world.countries where world.countries.capital = 'Abidjan'
```

```python
SELECT name
  FROM world.cities
  EXCEPT
SELECT capital
  FROM world.countries
ORDER BY name;
```

```python
-- Inversa

SELECT capital
FROM world.countries
EXCEPT
SELECT name
FROM world.cities
ORDER BY capital;
```
