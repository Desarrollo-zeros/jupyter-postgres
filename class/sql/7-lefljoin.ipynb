<h1>LEFT JOIN</h1>

- El Left e Inner Join traen mas registros que un Inner, dado que el inner descarta valores de la tabla de la derecha al no registrar coincidencias.  
- Un Left e Inner Join siempre traerá mas registros que un Inner Join, como se muestra a continuación:

<img src="https://www.w3schools.com/sql/img_leftjoin.gif"/>

## Ejemplo

```python
select * from world.countries limit 5
```

```python
select * from world.languages limit 5
```

Este consulta a continuación solo es para mostrar que un Left o Rigth joint trae mas registros que un Inner Join

```python
Select c.local_name AS country, l.name AS language, l.percent
FROM world.countries AS c
inner JOIN world.languages AS l
ON c.code = l.code
ORDER BY c.local_name DESC;
```

## Ahora si un Left Joint

```python
Select c.local_name AS country, l.name AS language, percent
FROM world.countries AS c
left JOIN world.languages AS l
ON c.code = l.code
ORDER BY c.local_name DESC;
```

```python
SELECT c.local_name, c.region, e.gdp_percapita
FROM world.countries AS c
LEFT JOIN world.economies AS e
ON c.code = e.code
WHERE e.year = 2010;
```

## Ejemplo

```python
SELECT region, AVG(gdp_percapita) AS avg_gdp
from world.countries as c
Left join world.economies as e
on c.code = e.code
where e.year = 2010
Group by c.region
order by avg_gdp desc;
```
