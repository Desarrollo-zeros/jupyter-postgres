<h1>FULL JOIN</h1>

El Full Join integra el left y right join, con el inner join, o sea, trae las intersecciones, los resultados del leff y del right.

<img src="https://www.tutorialrepublic.com/lib/images/full-join.png"/>

## Ejemplo

```python
select * from world.countries limit 5
```

```python
select * from world.currencies limit 5
```

```python
SELECT local_name AS country, code, region, basic_unit
FROM world.countries
FULL JOIN world.currencies
USING (code)
WHERE region = 'North America' OR region IS NULL
ORDER BY region;
```

```python
SELECT countries.local_name, code, languages.name AS language
FROM world.languages
FULL JOIN world.countries
USING (code)
WHERE countries.name LIKE 'V%' OR countries.name IS NULL
ORDER BY countries.name;
```
